## Explorando os dados do mdicos / Definio de funes de plot das imagens

def plot_images(df, start_index, num_images):


*   df =  qual o dataframe ser passado plotagem.
*   start_index =  a posio inicial do array.
*   num_images =  quantidade de imagens que sero plotadas








```python
#Plot o plot_images(DATA FRAME PARA PRINT, INDEX XDO DATA FRASE, QTD IMAGEM PARA PLOT)
def plot_images(df, start_index, num_images):
    fig = plt.figure(figsize=(10, 10))
    if start_index + num_images > len(df):
        num_images = len(df) - start_index
    for i in range(num_images):
        image_path = df['image'][start_index + i]
        image_class = df['class'][start_index + i] #Pegando a classe da imagem
        try:
            image = Image.open(image_path)
            ax = fig.add_subplot(num_images, 1, i+1)
            ax.imshow(image, cmap='gray')
            ax.set_title(f'Image {start_index + i}, Class: {image_class}')  #Imprimindo o titulo da classe da imagem
            ax.axis('off')
        #Excesso e tratamento de excesso em um acesso invalido.
        except (IOError, OSError) as e:
            print(f"Erro ao carregar imagem {start_index + i}: {e}")

    plt.tight_layout()
    plt.show()
```

# Preparao dos Dados para o Treinamento

Embora os modelos de aprendizado profundo sejam totalmente capazes de lidar com dados multidimensionais, como  o caso de imagens, a aplicao da operao de [reshape](https://www.tensorflow.org/api_docs/python/tf/reshape) para "achatar" as imagens e transform-las em uma nica matriz contnua de pixels  uma prtica comum. Essa operao simplifica a estrutura de dados de entrada e pode facilitar o processo de treinamento, especialmente ao lidar com modelos mais simples ou ao iniciar a modelagem.

* No nosso caso especfico, uma imagem de dimenses 28x28 pixels  transformada em uma matriz unidimensional de 784 elementos (28x28 = 784). Esta transformao  comumente referida como "nivelamento" da imagem.

* A funo `reshape`  uma ferramenta poderosa neste processo, permitindo que reformulemos a estrutura dos nossos dados para se adequar s necessidades do nosso modelo. A operao de reshape  crucial para garantir que os nossos dados de entrada estejam na forma correta antes de aliment-los no modelo.

Em nosso caso de classificao de imagens, temos duas categorias: 'Normal' e 'Pneumonia'. A ideia  que nosso modelo de aprendizado de mquina entenda que, para uma imagem especfica, ou ela  **'Normal' ou 'Pneumonia'**, no existe um "quase Normal" ou "quase Pneumonia".

O atual rtulo das imagens  uma designao de **'Normal' ou 'Pneumonia'**. Como essas categorias formam uma escolha binria, o modelo pode pensar erroneamente que est "prximo" de acertar se adivinhar de forma errada.

Para evitar isso, utilizaremos  a **codificao categrica**. Esta transformao muda os rtulos para que cada rtulo seja uma coleo de todas as categorias possveis, com o verdadeiro rtulo da imagem marcado como verdadeiro.

Por exemplo, para as categorias 'Normal' e 'Pneumonia', depois da codificao categrica, teramos algo assim para uma imagem 'Normal':

* 'Normal': Falso (1)
* 'Pneumonia': Verdadeiro (0)

Isso ajuda o modelo a entender que ou uma imagem  **'Normal' ou 'Pneumonia', sem qualquer conceito de proximidade entre as categorias**.

Esse processo  representado pela codificao das label `labels.append(1 if label == 'Pneumonia' else 0)

```python
# Funo para carregar e processar as imagens
def load_and_process_images(df, image_size=(28, 28)):
    images = []
    labels = []
    total_images = df.shape[0]
    for i, (image_path, label) in enumerate(zip(df['image'], df['class'])):
        # Carrega a imagem com a biblioteca PIL
        img = Image.open(image_path)
        #Redimensiona a imagem para 28x28 pixels usando o mtodo resize. (28 x 28)
        img = img.resize(image_size)
        #Converte a imagem para escala de cinza com o mtodo convert('L'). Isso  feito porque as imagens em escala de cinza so mais simples de lidar
        #(uma nica camada de intensidades de pixel, em vez de trs camadas de cores em imagens coloridas) e so suficientes para muitas tarefas de processamento
        #de imagem e aprendizado de mquina.
        img = img.convert('L')
        # Converte a imagem para um array NumPy
        img_array = np.array(img)
        # Normaliza os pixels (0-255) para o intervalo 0-1
        img_array = img_array / 255.0
        #Achatando a imagem usando reshape(-1). Isso transforma a matriz 2D em uma matriz 1D (um vetor) de elementos.
        #Como o tamanho da imagem  28x28 pixels, o vetor ter 784 elementos.
        img_array = img_array.reshape(-1)
        # Adiciona ao conjunto de imagens
        images.append(img_array)
        # Converte rtulo e adiciona ao conjunto de rtulos (Conerso de rotulos binria 1 Pneumonia 0 Normal)
        labels.append(1 if label == 'Pneumonia' else 0)

        # Mostrando o progresso
        if (i+1) % 100 == 0 or i+1 == total_images:
            print(f'Processando {i+1} de {total_images} imagens ({(i+1)/total_images*100:.2f}%)')

    return np.array(images), np.array(labels)
```

# Criao do Modelo

Com os dados prontos, construiremos um modelo bsico composto por camadas. O modelo tem 3 partes:
* Camada de entrada: recebe os dados no formato esperado.
* Camadas ocultas: formadas por neurnios que afetam a estimativa da rede atravs de seus pesos, valores ajustados ao longo do treinamento.
* Camada de sada: representa a estimativa da rede para uma entrada especfica.

Usamos a classe Dense do Keras para criar uma camada densamente conectada, onde cada neurnio afeta todos os da prxima camada. Definimos o nmero de neurnios com o argumento `units`. A funo de ativao `relu` permite que a rede faa suposies complexas sobre os dados. `input_shape` especifica o formato dos dados de entrada.

Por fim, adicionamos uma camada de sada com funo de ativao `softmax`, resultando em valores de probabilidade entre 0 e 1 que somam 1. Para a nossa tarefa de classificar uma imagem em 10 categorias possveis, teremos 10 sadas, cada uma representando a probabilidade da imagem pertencer a essa classe.

```python
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import Dropout

#reset da sesso e do modelo
tf.keras.backend.clear_session()

# Definindo o modelo
model = Sequential()

# Camada de entrada - Ns precisamos fazer a entrada de imagens em formato (altura, largura, canais)
# Para imagens em escala de cinza, a quantidade de canais  1. Se fossem imagens coloridas, seriam 3 (R, G, B)
model.add(Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=(28, 28, 1)))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))

model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))

model.add(Conv2D(128, (3, 3), padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))

model.add(Conv2D(256, (3, 3), padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))

model.add(Flatten()) # Achatamos as sadas 2D das camadas de convoluo para 1D para a entrada nas camadas densas

# Camada de entrada densa
model.add(Dense(1024, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.01)) # Reduzido a taxa de dropout

# Adicionar mais uma camada oculta para aumentar a complexidade do modelo
model.add(Dense(512, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.01)) # Reduzido a taxa de dropout

# Adicionar mais uma camada oculta para aumentar a complexidade do modelo
model.add(Dense(256, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.01)) # Reduzido a taxa de dropout

# Camada de sada
model.add(Dense(1, activation='sigmoid'))
```

# Treinando o Modelo

Com os dados de treinamento, validao e o modelo preparados,  hora de ajustar o modelo. Esse processo, tambm chamado de treinamento, adapta o modelo aos dados. Usamos o mtodo `fit` do Keras, que requer:

* Dados de treinamento e respectivos rtulos
* Nmero de iteraes completas no conjunto de treinamento (pocas)
* Dados de validao ou teste e seus rtulos

Ao executar o treinamento, o modelo aprender e ajustar sua forma para melhor representar os dados.

```python
from tensorflow.keras.callbacks import EarlyStopping
#se apos 10 epocs a taxa de perda no comear a diminuir o modelo ser interrompido
early_stopping = EarlyStopping(monitor='val_loss', patience=15)
#Treinamento do modelo
history = model.fit(x_train, y_train, batch_size=64, epochs=50, validation_data=(x_valid, y_valid), callbacks=[early_stopping])
```
